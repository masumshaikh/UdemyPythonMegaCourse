```python
import pandas
dt = pandas.read_csv('supermarkets.csv', index_col="ID")
dt
```

```python
# get row and column headers
dt.index
dt.columns
```

```python
# delete a column by name

# dt.drop('City') # doesn't work, axis is 0 by default and there is no row called 'City'
dt.drop('City', 1)
```

```python
# delete a row by index
dt.drop(2) # works because axis is 0 by default
```

```python
# delete more than one column in one go
dt.drop(dt.columns[0:3], 1)
```

```python
# delete non-contiguous columns in one go
dt.drop(dt[['Country', 'Name']],1)

# note that to extract the two columns from dt,
# you do NOT do something like dt.columns[...]
```
